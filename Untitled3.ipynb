```python
# prompt: now train a cnn model based on the 2 folders i.e test which contains 1 folder of cataract images and another of normal eye images and the train folder consists of a cataract folder and a normal folder

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define image dimensions and batch size
img_width, img_height = 150, 150
batch_size = 32

# Create data generators for training and testing data
train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        '/content/processed_images/train',
        target_size=(img_width, img_height),
        batch_size=batch_size,
        class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
        '/content/processed_images/test',
        target_size=(img_width, img_height),
        batch_size=batch_size,
        class_mode='binary')


# Build the CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(img_width, img_height, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // batch_size,
        epochs=10,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // batch_size)

# Save the trained model
model.save('cataract_detection_model.h5')
```

```python
# prompt: train vgg 16 to classify whether cataract is dectected or not there are 2 folders a test and train folder and create directories to separately access both the folders and its contents i.e under test folder make seperate directories for cataract and normal and under train make seperate directories for cataract and normal the model should be able to detect cataract

import os
import shutil
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create directories for cataract and normal images in train and test folders
def create_directories(base_dir):
  if not os.path.exists(base_dir):
    os.makedirs(base_dir)

  for sub_dir in ['cataract', 'normal']:
    dir_path = os.path.join(base_dir, sub_dir)
    if not os.path.exists(dir_path):
      os.makedirs(dir_path)

create_directories('/content/processed_images/train')
create_directories('/content/processed_images/test')


# Move images to the appropriate folders
def move_images(source_dir, destination_dir):
  for filename in os.listdir(source_dir):
    if 'cataract' in filename:
      shutil.move(os.path.join(source_dir, filename), os.path.join(destination_dir, 'cataract'))
    elif 'normal' in filename:
      shutil.move(os.path.join(source_dir, filename), os.path.join(destination_dir, 'normal'))

# Move images from the original train and test folder to the created folders
# if you have a different folder structure, adjust the paths
move_images('/content/processed_images/train', '/content/processed_images/train')
move_images('/content/processed_images/test', '/content/processed_images/test')




# Define image dimensions and batch size
img_width, img_height = 224, 224
batch_size = 32

# Create data generators for training and testing data
train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        '/content/processed_images/train',
        target_size=(img_width, img_height),
        batch_size=batch_size,
        class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
        '/content/processed_images/test',
        target_size=(img_width, img_height),
        batch_size=batch_size,
        class_mode='binary')


# Load the VGG16 model without the top layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(img_width, img_height, 3))

# Add new layers for binary classification
x = Flatten()(base_model.output)
x = Dense(128, activation='relu')(x)
predictions = Dense(1, activation='sigmoid')(x)

# Create the new model
model = Model(inputs=base_model.input, outputs=predictions)

# Freeze the layers of the VGG16 model
for layer in base_model.layers:
    layer.trainable = False

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // batch_size,
        epochs=15,  # You can adjust the number of epochs
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // batch_size)

# Save the trained model
model.save('cataract_detection_vgg16_model.h5')

```

```python
# prompt: create interface and predict whether cataract or not and display image

from google.colab import files
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import io

# Load the trained model
model = load_model('/content/cataract_detection_vgg16_model.h5')  # Replace with your model file name
model0 = load_model('/content/cataract_detection_model.h5')  # Replace with your model file name

def predict_cataract(uploaded_file):
  """Predicts whether an uploaded image has cataract or not."""

  image_bytes = uploaded_file.read()
  image_data = io.BytesIO(image_bytes)
  img = image.load_img(image_data, target_size=(224, 224))
  x = image.img_to_array(img)
  x = np.expand_dims(x, axis=0)
  x = x / 255.0  # Normalize the image

  prediction = model.predict(x)
  pred_0=model0.predict(x)

  if prediction[0][0] < 0.5:
    result = "Cataract Detected "
  else:
    result = "No Cataract Detected"

  return result, img


def display_prediction():
  """Displays an uploaded image and its prediction."""

  uploaded = files.upload()

  for fn in uploaded.keys():
    uploaded_file = io.BytesIO(uploaded[fn])
    result, img = predict_cataract(uploaded_file)

    plt.imshow(img)
    plt.title(result)
    plt.axis('off')
    plt.show()


# Run the prediction interface
display_prediction()
```
